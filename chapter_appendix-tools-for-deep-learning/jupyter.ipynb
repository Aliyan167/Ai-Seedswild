```python
def jacobi_method(error_tolerance=1e-4):
    # Initial guesses
    x, y, z = 0, 0, 0
    count = 0

    while True:
        count += 1
        # Update using previous values
        x_new = (1 - 3*y - z) / 2
        y_new = (2 - 4*x + 2*z) / 1
        z_new = (7 - 3*x - 2*y) / 3

        # Compute error (max of absolute differences)
        error = max(abs(x_new - x), abs(y_new - y), abs(z_new - z))

        print(f"Iteration {count}: x={x_new:.6f}, y={y_new:.6f}, z={z_new:.6f}, error={error:.6f}")

        if error < error_tolerance:
            break

        # Update old values
        x, y, z = x_new, y_new, z_new

    print("\nFinal Approximate Solution (Jacobi Method):")
    print(f"x = {x_new:.6f}, y = {y_new:.6f}, z = {z_new:.6f}")
```
